# 02 — Strength Model & Passport Generation

**AI-Enabled Concrete Performance Passports for OpenBIM**

Runnable in Google Colab. Installs dependencies, downloads the UCI Concrete Compressive Strength dataset, trains a compressive-strength model, evaluates it, generates explanations, and builds Concrete Performance Passports.

> Phase-6 deliverable; populated during implementation. Backed by `src/` modules.

## What this notebook will produce
- Trained strength model + metrics (R2, RMSE, MAE, MAPE).
- Parity, residual, and feature-importance figures in `outputs/figures/`.
- Concrete Performance Passports in `outputs/passports/`.
