# 03 — Carbon & OpenBIM Mapping

**AI-Enabled Concrete Performance Passports for OpenBIM**

Runnable in Google Colab. Estimates embodied carbon (cradle-to-gate GWP) per mix from composition using published per-constituent factors, derives carbon classes/benchmarks, and produces the OpenBIM/IFC mapping plus an IDS-ready requirement matrix.

> Phase-6 deliverable; populated during implementation. Backed by `src/train_carbon_model.py` and `src/ifc_mapping.py`.

## What this notebook will produce
- GWP estimates + carbon classes (Low/Typical/High) with benchmark percentiles.
- GWP-vs-strength figure in `outputs/figures/`.
- Passport-to-IFC mapping table, proposed `Pset_ConcretePerformancePassport`, and IDS-ready requirement matrix (CSV + conceptual IDS XML).
